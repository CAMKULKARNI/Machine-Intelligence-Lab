```python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import *
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("D:\\codes_python\\MI Lab\\week 6\\train.csv")
X = data.iloc[:, 0:-1]
y = data.iloc[:, -1]
```

```python
data = pd.read_csv('test.csv')
X_test = data.iloc[:, 0:-1]
y_test = data.iloc[:, -1]
```

```python
for scaler in [StandardScaler(), MaxAbsScaler(), MinMaxScaler(), RobustScaler(), PowerTransformer()]:
    for kernel in ['rbf', 'linear', 'poly', 'sigmoid']:
        classifier = SVC(kernel=kernel)
        pipe = Pipeline([('scaler', scaler), ('svc', classifier)])
        pipe.fit(X, y)
        print(f'Accuracy: {pipe.score(X_test, y_test)*100:.2f}%', scaler, kernel)
```

```python
for gamma in ['auto', 'scale'] + [i for i in range(1, 11)]:
    classifier = SVC(kernel='rbf', gamma=gamma)
    pipe = Pipeline([('scaler', StandardScaler()), ('svc', classifier)])
    pipe.fit(X, y)
    print(f'Accuracy: {pipe.score(X_test, y_test)*100:.2f}%')
```

```python
for scaler in [StandardScaler(), MaxAbsScaler(), MinMaxScaler(), RobustScaler(), PowerTransformer()]:
    for degree in range(1, 11):
        for coef0 in range(1, 11):
            classifier = SVC(kernel='poly', degree=degree, coef0=coef0)
            pipe = Pipeline([('scaler', scaler), ('svc', classifier)])
            pipe.fit(X, y)
            print(f'Accuracy: {pipe.score(X_test, y_test)*100:.2f}%', scaler, degree, coef0)
```
